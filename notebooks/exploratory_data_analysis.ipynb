```python
# üõ†Ô∏è Create an instance of DataPreprocessor
processor = DataPreprocessor()
analyzer = StockDataExploration()
```

```python
bnd_data = processor.load_data('BND')
spy_data = processor.load_data('SPY')
tsla_data = processor.load_data('TSLA')
```

```python
# Define dictionary for stock data for TSLA, BND, SPY
data_dict = {
    'BND': bnd_data,    
    'SPY': spy_data,  
    'TSLA': tsla_data 
}

# Print statement with emojis for better visualization
print("üìà Stock Data Dictionary Created Successfully! üöÄ")
print("Stocks Included:")
for symbol in data_dict.keys():
    print(f" ‚úÖ {symbol} üìä")
```

```python
# üìà Plotting the closing price over time 
print("üìä Preparing to Plot Closing Prices Over Time... ‚è≥")
analyzer.plot_closing_price(data_dict)
print("‚úÖ Closing Prices Plotted Successfully! üéâ")
```

```python
# üìà Plotting the daily percentage change in the closing price to observe volatility.
print("üìä Preparing to Plot Daily Percentage Change in Closing Prices... ‚è≥")
analyzer.plot_percentage_change(data_dict)
print("‚úÖ Daily Percentage Change Plotted Successfully! üéâ")
```

```python
# üìà Plotting the closing price, rolling mean, and volatility (rolling std) over time for multiple symbols.
print("üìä Starting Analysis of Price Trends... ‚è≥")
analyzer.analyze_price_trend(data_dict)
print("‚úÖ Price Trend Analysis Completed Successfully! üéâ")
```

```python
# üîç Detecting outliers in the stock data
print("üìä Initiating Outlier Detection... ‚è≥")
analyzer.detect_outliers(data_dict)
print("‚úÖ Outlier Detection Completed Successfully! üéâ")
```

```python
# üìà Plotting the closing price for a stock symbol (e.g., 'TSLA')
stock_symbol = 'TSLA'  # Specify the stock symbol you want to plot
print(f"üìä Preparing to Plot Daily Percentage Change for {stock_symbol}... ‚è≥")
analyzer.plot_percentage_change({stock_symbol: data_dict[stock_symbol]})
print("‚úÖ Daily Percentage Change Plotted Successfully! üéâ")
```

```python
# üìà Plotting unusual daily returns for the stock data
print("üìä Initiating Plot of Unusual Daily Returns... ‚è≥")
analyzer.plot_unusual_daily_return(data_dict)
print("‚úÖ Unusual Daily Returns Plotted Successfully! üéâ")
```
